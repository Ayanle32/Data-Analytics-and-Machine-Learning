```python
"""
The data sets wer are using come from sklearn.  We need matplotlib for plotting out scatter plots and histograms,
and we will be using numpy to compute covaraince and correlation matrices. So all three are imported at the
start of this notebook.
"""
print("Import the sklearn datasets")
import sklearn.datasets as ds
import matplotlib.pyplot as plt
import numpy as np

"""
The Iris data is to be used for the default study.  There are other suggested data sets that can be accessed
by commenting out the Iris data lines, and uncommenting the corresponding lines for one of the other data sets.
"""
#print("use the Boston house price data")
#Bunch = ds.load_boston(return_X_y=False) # we want to get the Bunch so that we can access more than just the data and target values
#target_name = "MEDV"
#target_plot_xaxis = "target value (house prices, MEDVAL $k US)"

#print("use the brest cancer data")
#Bunch = ds.load_breast_cancer(return_X_y=False) # we want to get the Bunch so that we can access more than just the data and target values
#target_name = "disease progression"
#target_plot_xaxis = "target value (Malignant=0, Benign=1)"

#print("use the diabetes data")
#Bunch = ds.load_diabetes(return_X_y=False) # we want to get the Bunch so that we can access more than just the data and target values
#target_name = "disease progression"
#target_plot_xaxis = "target value (disease progression)"

print("use the iris data")
Bunch = ds.load_iris(return_X_y=False) # we want to get the Bunch so that we can access more than just the data and target values
target_name = "class"
target_plot_xaxis = "target value (class)"

#print("use the linnerud(exercise) data")
#Bunch = ds.load_linnerud(return_X_y=False) # we want to get the Bunch so that we can access more than just the data and target values
#target_name = "class"
#target_plot_xaxis = "target value (class)"

#print("use the wine data")
#Bunch = ds.load_wine(return_X_y=False) # we want to get the Bunch so that we can access more than just the data and target values
#target_name = "class"
#target_plot_xaxis = "target value (class)"


# Print the description of the data
print(Bunch.DESCR)
#print("The following data file is being used: ", Bunch.filename, "\n")

# no need to print the feature names as these are printed out with DESCR 
# print(Bunch.feature_names)

# un-comment hte following lines to print out the data and the target values.
# 
# print(Bunch.data)
# print(Bunch.target)

i=0
for this_feature in Bunch.feature_names:
    print("Working on feature", this_feature)
    plt.hist(Bunch.data[:,i], 10)
    plt.xlabel(this_feature)
    plt.ylabel("Number of entries")
    plt.show()
    i+=1   # increment the counter to acces the next feature

print("Working on the target values")
plt.hist(Bunch.target, 10)
plt.xlabel(target_plot_xaxis) # use the value set above when selecting the data set
plt.ylabel("Number of entries")
plt.show()
```

```python
"""
Compute the mean and standard deviation of the data set, formatted to a certain
number of decimal places. Note that some data sets will have this information 
as a part of the description printed out as a result of runnign the previous code
cell
"""
print("-------------------------------------------------------------")
print("Variable       \t        Average \t St Dev ")
print("-------------------------------------------------------------")

for i in range(len(Bunch.feature_names)):
    print("{0:15s} \t {1:5.3f}    \t {2:5.3f}".format(Bunch.feature_names[i], np.average(Bunch.data[:,i]), np.std(Bunch.data[:,i])))
print("{0:15s} \t {1:5.3f}    \t {2:5.3f}".format(target_name, np.average(Bunch.target), np.std(Bunch.target)))

print("-------------------------------------------------------------")


"""
Now compute the covariance matrix and print that out.  Note that the sklearn data samples
use rows for examples and columns for features. The default numpy correlation calculation
is to use rows for features and columns for examples; so we have to set rowvar=False to
reverse that default ordering to get the correct matrix.
"""
cov_matrix = np.cov(Bunch.data, rowvar=False)
print("\nThe covariance matrix for these data is:\n")
print(cov_matrix)

"""
Also compute the correlation matrix; again need to use the rowvar=False option
"""
corr_matrix = np.corrcoef(Bunch.data, rowvar=False)
print("\nThe correlation matrix for these data is:\n")
print(corr_matrix)
```
