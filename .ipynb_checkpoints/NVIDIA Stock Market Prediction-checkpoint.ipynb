# Abstract

In this project we will harness the power of machine learning to predict the stock price of Nvidia and the S&P 500. Since the stock market is highly chaotic and dependent on external factors such as global events this prediction only aims to detect temporal patterns within the raw data itself. A much more accurate prediction will require the consideration of these factors.

# Train Test Split & Data Preprocessing

```python
X = df[['Year', 'Month', 'Day']]
y = df['Close'].values # target variable
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 21) # splitting the dataset
```

```python
X2 = df2[['Year', 'Month', 'Day']]
y2 = df2['Close'].values # target variable
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size = 0.2, random_state = 21)
```

# Fitting Random Forest Regressor & Predictions

```python
model = RandomForestRegressor(random_state = 21) # initialising the rf model
model.fit(X_train, y_train) # fitting the model
model2 = RandomForestRegressor(random_state = 21)
y_pred = model.predict(X_test)
y_train_pred = model.predict(X_train)
```

```python
model2.fit(X2_train, y2_train) # fitting the model
y2_pred = model2.predict(X2_test)
y_train_pred2 = model2.predict(X2_train)
```

```python
future_dates = pd.date_range(start=df['Date'].iloc[-1], periods=60)
future_dates2 = pd.date_range(start=df2['Date'].iloc[-1], periods=60)
future_features = pd.DataFrame({'Year': future_dates.year,
                                'Month': future_dates.month,
                                'Day': future_dates.day})
future_features2 = pd.DataFrame({'Year': future_dates2.year,
                                'Month': future_dates2.month,
                                'Day': future_dates2.day})
future_predictions = model.predict(future_features) # NVIDIA
future_predictions2 = model2.predict(future_features2)
```

# Model Performance

```python
test_rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(test_rmse)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
```

```python
param_grid_rf = {
    'n_estimators': [50, 100],
    'max_depth': [None, 10],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

param_grid_dt = {
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

param_grid_gbm = {
    'n_estimators': [50, 100],
    'learning_rate': [0.05, 0.1, 0.2],
    'max_depth': [3, 5, 7]
}

param_grid_xgb = {
    'n_estimators': [50, 100],
    'learning_rate': [0.05, 0.1, 0.2],
    'max_depth': [3, 5, 7]
}
# Create dictionary of models and their respective parameter grids
models = {
    'RandomForest': (RandomForestRegressor(), param_grid_rf),
    'DecisionTree': (DecisionTreeRegressor(), param_grid_dt),
    'GradientBoosting': (GradientBoostingRegressor(), param_grid_gbm),
    'XGBoost': (XGBRegressor(), param_grid_xgb)
    
}

# Perform grid search for each model

scoring = 'neg_mean_squared_error'

for model_name, (model, param_grid) in models.items():
    print(f"Grid search for {model_name}...")
    grid_search = GridSearchCV(model, param_grid, cv=5, scoring=scoring)
    grid_search.fit(X_train, y_train)
    
    # Get the best hyperparameters
    best_params = grid_search.best_params_
    print(f"Best hyperparameters for {model_name}: {best_params}")
    
    # Evaluate the best model on the test set
    best_model = grid_search.best_estimator_
    y_pred = best_model.predict(X_test)
    score = grid_search.best_score_
    print(f"Accuracy of the best {model_name} model on test set: {score:.4f}\n")
```

```python
xgb = XGBRegressor(learning_rate = 0.2, max_depth = 7, n_estimators = 100)
xgb2 = XGBRegressor(learning_rate = 0.2, max_depth = 7, n_estimators = 100)
```

```python
xgb.fit(X_train, y_train)
xgb2.fit(X2_train, y2_train)
```

```python
future_predictions3 = xgb.predict(future_features)
future_predictions4 = xgb2.predict(future_features2)
```
