```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

# Loading in the Dataset

```python
df = pd.read_csv('WeatherDataM.csv')
df.head()
```

So here we have some weather information and our goal is to determine whether there is a linear relationship between humidity and other features.

# Data Visualisation

```python
plt.figure(figsize=(20, 10))
plt.subplot(3, 3, 1)
plt.scatter(df['Temperature (C)'], df['Humidity'], s=10, color='blue')
plt.xlabel('Temperature (C)')
plt.ylabel('Humidity')
plt.subplot(3,3,2)
plt.scatter(df['Wind Speed (km/h)'], df['Humidity'], s=10, color = 'blue')
plt.xlabel('Wind Speed (km/h)')
plt.ylabel('Humidity')
plt.subplot(3,3,3)
plt.scatter(df['Pressure (millibars)'], df['Humidity'], s=10, color = 'blue')
plt.xlabel('Pressure (millibars)')
plt.ylabel('Humidity')
plt.show()
```

Immediately we can disregard pressure in our linear model because from the scatter plot we can see it does not follow a linear relationship with humidity.

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection = "3d")

ax.scatter(df['Temperature (C)'], df['Wind Speed (km/h)'], df['Humidity'])
ax.set_xlabel('Temperature (C)')
ax.set_ylabel('Wind Speed (km/h)')
ax.set_zlabel('Humidity')
plt.show()
```

# Train-Test Split

```python
X = df.drop(columns = ['Pressure (millibars)', 'Humidity'])
y = df['Humidity']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 21)
```

# Fitting the Model

```python
LR = LinearRegression()
LR = LR.fit(X_train, y_train)
```

```python
LR.coef_ 
```

```python
LR.intercept_
```

We have determined the parameters of the model and we can make predictions on new data.
