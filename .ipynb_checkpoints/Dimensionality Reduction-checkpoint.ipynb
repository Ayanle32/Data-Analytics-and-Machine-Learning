When dealing with large datasets we usually have thousands or even millions of features for each training instance. This can make training extremely slow. Therefore, it is crucial to reduce the dimensions of the data. For example in the MNIST dataset we could drop the pixels at the image border since they are all the same colour and don't do much for the classification task. 
We could also merge pixels by taking the average of their intensities. Both of these methods reduce the dimensions of the data.
Dimensionality reduction can also be important when trying to visualise data especially when trying to detect patterns or when presenting insights to non-technical stakeholders.

As the number of dimensions increases the distance between training instances in the feature space also increases. Certain machine learning models struggle in high dimensional datasets such as  nearest neighbours and k means as these rely on distance metrics. As the ratio of features to instances increases high the risk of overfitting increases. This is because when the number of features are high the model has more parameters to estimate and it begins to capture noise and fluctuations in the data.

# Approaches to Dimensionality Reduction

## Projection

If many features are correlated or constant it is possible to project the training instances to a lower dimensional subspace

## Manifold Learning

A manifold is a topological space that locally resembles Euclidean space. A sheet of paper is a 2D manifold because it locally looks like a 2D Euclidean space. However, if twisted in a 3 dimensional space it becomes a 2D manifold embedded in 3 dimensions. Similarly, the surface of a sphere is a 2D manifold embedded in 3 dimensional space.

Manifold Learning is a technique which aims to reduce the dimensionality of the data by finding a lower-dimensional manifold that can capture the data within the higher dimensional space. Manifold Learning algorithms aim to find this manifold and provide a low-dimensional representation of the data while preserving certain properties of the original high-dimensional data.

A caveat with dimensional reduction is that reducing the dimensionality of the data does not necessarily lead to better a solution although it does speed up training.

# PCA

Principle Component Analysis finds the hyperplane that lies closest to the data points and project. The hyperplane that preserves the maximum amount of variances is chosen since it is the axis that minimises the mean squared distance between the 
original dataset and its projection onto that axis.

## Principle Components

PCA identifies the axis that accounts for the largest variance in the training set. It also finds a second axis orthogonal to the first axis that accounts for the remaining variances. In D-dimensions it finds D such axis. 
Each axis is defined by a unit vector called the principle component. The principle components are found by factorising the centred training set matrix $\bf{X}_c$. The factorisation technique used is Singular Value Decomposition (SVD) $X_c = U\Sigma V^T$.

The covariance matrix is 

$C = \frac{1}{n-1} X_{c}^TX_c$.

Substituting the decomposition gives 

$ C = \frac{1}{n-1}\left(V\Sigma U^T)^T(U\Sigma V^T \right)$

using $ UU^T = I_{n}$
and   $ \Sigma ^T \Sigma = \Lambda$

$ C = V \Lambda V^T $

So the right singular vectors $V$ contain the eigenvectors of $C$ and $\Lambda$ contain the eigenvalues.

The dimensions of the data are then reduced 

$X_{reduced} = X_c V_k$.


## Finding the Principle Components

```python
import numpy as np
from sklearn.datasets import load_iris
```

```python
X = load_iris().data
y = load_iris().target
```

```python
X_c = X - X.mean(axis=0)
U, s, Vt = np.linalg.svd(X_c)
```

```python
c1 = Vt.T[:, 0] # 1st principle component which is the direction in the feature space along which the variance is maximised.
c2 = Vt.T[:, 1]
c3 = Vt.T[:, 2]
c4 = Vt.T[:, 3]
```

```python
c1
```

```python
s # Magnitudes of the variances in the data along the principle components
```

## Projecting Down to d Dimensions

$ X_{d-proj} = XW_d$

```python
W2 = Vt.T[:, :2] # first two principle components
X2D = X_c.dot(W2) # data projected to a 2D plane
```

```python
import matplotlib.pyplot as plt
plt.scatter(X2D[:, 0], X2D[:, 1])
plt.show()
```

## Choosing the Right Number of Dimensions

A good way to choose the number of dimensions is to choose the number of dimensions that add up to a sufficiently large portion of the variance for example 90%.

```python
pca = PCA()
pca.fit(X)
cumsum = np.cumsum(pca.explained_variance_ratio_)
d = np.argmax(cumsum >= 0.95) + 1
d
```

```python
pca = PCA(n_components = 0.95) # Does the same thing as the cell above and is much easier
```

## PCA for Compression

```python
pca = PCA(n_components = 200) # data reduced to 200 dimensions
X_reduced = pca.fit_transform(X)
X_decompressed = pca.inverse_transform(X_reduced) # data decompressed to original number of dimensions
```

The decompressed data is not identical to the original since the projection lost some information. The mean squared distance between the original data and the decompressed data is the reconstruction error.

The equation for the inverse PCA transformation is

$X_{decompressed} = X_{reduced}V^T$.

```python
import matplotlib as mpl
digit = np.array(X.iloc[4, ]).reshape(28, 28)
plt.figure(figsize=(10, 6))
plt.subplot(1,2, 1)
plt.title('Original', fontsize=16)
plt.imshow(digit, cmap = mpl.cm.binary)
plt.axis('off')
plt.subplot(1,2, 2)
plt.title('Decompressed Data (200 dimensions)', fontsize=16)
plt.imshow(X_decompressed[4,].reshape(28,28), cmap = mpl.cm.binary)
plt.axis('off')
plt.show()
```

## Randomized PCA

By setting the svd_solver hyperparameter to "randomized" we can force Scikit-Learn to use a stochastic PCA algorithm that finds an approximation of the first d principle components. This has a smaller time complexity and is quicker.

```python
rand_pca = PCA(n_components=100, svd_solver = 'randomized')
X_red = rand_pca.fit_transform(X)
```

By default Scikit-Learn uses svd_solver = auto which uses randomised PCA if m,n > 500 and d < 80% of m or n. We can force Scikit-Learn to use full SVD by setting svd_solver = full.

## Incremental PCA

Incremental PCA works by splitting the training set into mini batches and feeding these mini batches into the Incremental PCA algorithm one at a time. It is useful when the training set is too large to fit in memory. It is also useful for online learning as it can update the principle components as new instances arrive.

```python
from sklearn.decomposition import IncrementalPCA
```

```python
n_batches = 50
ipca = IncrementalPCA(n_components = 2)

for batch in np.array_split(X, n_batches):
    ipca.partial_fit(batch)
```

```python
X_red = ipca.transform(X)
```

## Kernel PCA

Kernel PCA makes it possible to perform complex nonlinear projections for dimensionality reduction. Kernel PCA is good at preserving clusters of instances and unrolling datasets that lie close to a twisted manifold.

```python
from sklearn.datasets import make_swiss_roll
```

```python
swissroll = make_swiss_roll(n_samples = 2000, noise = 0)
```

```python
plt.figure(figsize = (15, 9))
ax = plt.axes(projection='3d')
ax.scatter(swissroll[0][:, 0], swissroll[0][:, 1], swissroll[0][:, 2], c = swissroll[1], cmap = 'hot')
plt.show()
```

```python
from sklearn.decomposition import KernelPCA
```

```python
lin_pca = PCA(n_components = 2)
X_reduced = lin_pca.fit_transform(swissroll[0])
```

```python
rbf_kpca = KernelPCA(n_components = 2, kernel = 'rbf', gamma=0.04)
X_rbf = rbf_kpca.fit_transform(swissroll[0])
plt.figure(figsize=(6, 4))
plt.scatter(X_rbf[:, 0], X_rbf[:, 1], c=swissroll[1], cmap = 'hot')
plt.title('RBF Kernel, $\gamma = 0.04$', fontsize=16)
plt.xlabel('$Z_1$', fontsize=14)
plt.ylabel('$Z_1$', rotation = 360, labelpad = 10, fontsize=14)
plt.grid()
```

```python
sigmoid_kpca = KernelPCA(n_components = 2, kernel = 'sigmoid', gamma=0.001, coef0 = 1)
X_red = sigmoid_kpca.fit_transform(swissroll[0])
plt.figure(figsize=(6, 4))
plt.scatter(X_red[:, 0], X_red[:, 1], c=swissroll[1], cmap = 'hot')
plt.grid()
plt.title('Sigmoid Kernel, $\gamma = 0.001$, r = 1', fontsize=16)
plt.xlabel('$Z_1$', fontsize=14)
plt.ylabel('$Z_1$', rotation = 360, labelpad = 10, fontsize=14)
plt.grid()
```

## Selecting a Kernel & Tuning Hyperparameters

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
```

```python
clf = Pipeline([
    ('kpca', KernelPCA(n_components=2)),
    ('clf', LogisticRegression())
])

param_grid = [{
    'kpca__gamma': np.linspace(0.03, 0.05, 10),
     'kpca__kernel': ['rbf', 'sigmoid'], 
    'clf__C': [0.001, 0.01],
    'clf__max_iter': [500, 1000]
}]
```

```python
X = load_iris().data
y = (load_iris().target == 0).astype(int)
```

```python
grid_search = GridSearchCV(clf, param_grid, cv=5, scoring='accuracy')
```

```python
grid_search.fit(X, y)
```

```python
grid_search.best_params_
```

```python
np.max(grid_search.cv_results_['mean_test_score']) 
```

Alternatively, we can use an unsupervised approach where we select the kernel and hyperparameters that yield the lowest reconstruction error.

```python
rbf_pca = KernelPCA(n_components = 2, kernel = 'rbf', gamma = 0.03, fit_inverse_transform=True)
X_reduced = rbf_pca.fit_transform(X)
X_preimage = rbf_pca.inverse_transform(X_reduced)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
recon_error = mean_squared_error(X, X_preimage)
recon_error
```

# Locally Linear Embedding (LLE)

LLE is a non-linear dimensionality reduction technique. As it is a manifold learning technique it does not rely on projections like the previous techniques. LLE works by measuring how each training instance linearly relates to its closest neighbours and then the algorithm looks for a lower dimensional representation where these local relationships are preserved.

```python
from sklearn.manifold import LocallyLinearEmbedding
```

```python
swissroll = make_swiss_roll(n_samples = 2000, noise = 0)
plt.figure(figsize=(25, 10))
ax = plt.axes(projection = '3d')
ax.scatter(swissroll[0][:, 0], swissroll[0][:, 1], swissroll[0][:, 2], c = swissroll[1], cmap = 'hot', s = 40)
plt.title('Swissroll', fontsize=16)
plt.show()
```

```python
X = swissroll[0]
lle = LocallyLinearEmbedding(n_components = 2, n_neighbors = 10)
X_reduced = lle.fit_transform(X)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X_reduced[:,0], X_reduced[:, 1], c = swissroll[1], cmap = 'hot')
plt.title('Unrolled Swiss Roll Using LLE', fontsize=18)
plt.xlabel('$Z_1$', fontsize=16)
plt.ylabel('$Z_2$', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid()
plt.show()
```

On this plot we can see distances between instances are locally well preserved.

## How LLE Works Under the Hood

For each training instance $x^i$, the algorithm identifies its k nearest neighbours and then tries to express $x^i$ as a linear combination of these neighbours. The weights $w_{i,j}$ are found by minimising the distance between $x^i$ (some instance) and 
$\sum_{j=1}^{m} w_{i,j}x^{(j)}$ (a linear combination of the k nearest neighbours). Here we make the assumption that $w_{i,j} = 0$ if $x^{j}$ is not one of the k nearest neighbours of x^i.

$\hat{W} = argmin \sum^{m}_{i = 1}\left(x^i - \sum^{m}_{j = 1} w_{i,j}x^j\right)^2$

This is subject to 

$w_{i,j} = 0$ if $x^{j}$ is not one of the k nearest neighbours of x^i. $\newline$
$\sum^{m}_{j=1}w_{i, j} = 1$ for $ i = 1,2,\cdots,m$. The weights are normalised.

At this point we have the weight matrix which encodes the local linear relationships between the training instances. The next step is to map the training instances into a dimensional space with $d<n$. If $\bf{z}^i$ is the image of $\bf{x}^i$ then the goal is to minimise

$\hat{Z} = argmin \sum_{i = 1}^{m}\left(z^i - \sum_{j = 1}^{m} \hat{w}_{i,j} z^j\right)^2$.

Here we using the weights we found to find the optimal position of the instances' images in the low-dimensional space.

## Computational Complexity

Finding k nearest neighbours - $\mathcal{O}\left(m\log(m)n\log(k)\right)$ $\newline$
Optimising the weights - $\mathcal{O}\left(mnk^3\right)$ $\newline$
Constructing the low-dimensional representation - $\mathcal{O}\left(dm^2\right)$
The $m^2$ in the last term makes this extremely slow for large datasets
