```python
import numpy as np
from datetime import datetime
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = np.loadtxt('facebook.csv', dtype='object', delimiter=',', unpack=True, skiprows=1)
```

```python
dates = data[0]
dates = dates.astype(str)
high = data[2]
high = high.astype(float)
low = data[3]
low = low.astype(float)
Open = data[1]
close = data[4]
Open = Open.astype(float)
close = close.astype(float)
```

```python
dates = pd.to_datetime(dates)
```

```python
plt.figure(dpi=108, figsize=(10,6))
plt.title("High and low stocks for Facebook from 20/06/2019-19/06/2020")
plt.xlabel('')
plt.ylabel('Stock Price($)', fontsize='12')
plt.tick_params(axis='both', which='major', labelsize=12)
plt.plot(dates, high, c='red', label='High')
plt.plot(dates, low, c='blue', label='Low')
plt.fill_between(dates,high,low, facecolor='blue', alpha=0.1)
plt.legend()
plt.show()
```

```python
plt.figure(dpi=108, figsize=(10,6))
plt.title("Opening and closing stocks for Facebook from 20/06/2019-19/06/2020")
plt.xlabel('')
plt.ylabel('Stock Price($)', fontsize='12')
plt.tick_params(axis='both', which='major', labelsize=12)
plt.plot(dates, Open, c='green', label='Opening Stock')
plt.plot(dates, close, c='purple', label='Closing Stock')
plt.fill_between(dates,Open,close, facecolor='blue', alpha=0.1)
plt.legend()
plt.show()
```
